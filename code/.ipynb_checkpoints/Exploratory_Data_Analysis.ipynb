```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Load datasets
seattle_rain = pd.read_csv("../data/seattle_rain.csv")
stl_rain = pd.read_csv("../data/stl_rain.csv")

# Add city column
seattle_rain["city"] = "Seattle"
stl_rain["city"] = "St. Louis"

# Rename PRCP to precipitation
seattle_rain = seattle_rain.rename(columns={"PRCP": "precipitation"})
stl_rain = stl_rain.rename(columns={"PRCP": "precipitation"})

# Combine
df = pd.concat([seattle_rain, stl_rain], ignore_index=True)

# Convert DATE to datetime
df["DATE"] = pd.to_datetime(df["DATE"], errors="coerce")

# Extract month
df["Month"] = df["DATE"].dt.month

# Group by city + month and compute mean precipitation
monthly = df.groupby(["city", "Month"], as_index=False)["precipitation"].mean()

# Month names for x-axis
month_names = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]
```

```python
#PLOT THE DAILY PRECIPITATION FOR BOTH CITIES ON THE SAME GRAPH

plt.figure(figsize=(20,5))

sns.lineplot(data=df, x= 'DATE', y='precipitation', hue='city')

plt.xlabel('DATE', fontsize=15)
plt.ylabel('Daily precipitation (inches)', fontsize=15)

plt.tick_params(labelsize=15)

plt.show()
```

ST.louis has more precipitation on the daily than Seattle.

```python
df = pd.read_csv('https://raw.githubusercontent.com/brian-fischer/DATA-5100/refs/heads/main/clean_seattle_stl_weather.csv')
```

```python
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
#perform a statistical test for differences in the mean precipitation each month between the cities

plt.figure(figsize=(5,4))

sns.histplot(data=df.loc[df['month'] == 1], x='precipitation', hue='city')

plt.xlabel('Precipitation (inches)', fontsize=20)
plt.ylabel('Number of days', fontsize=20)
plt.title('Precipitation in January', fontsize=20)

plt.tick_params(labelsize=15)

plt.show()
```

```python
from scipy import stats
```

```python
from statsmodels.stats.proportion import proportions_ztest

significance_level = 0.05
significance_difference = np.zeros(12) 

# Create a Boolean column: True if precipitation > 0
df["any_precipitation"] = df["precipitation"] > 0

# Perform t-test for each month
for month in range(1, 13):

    contingency_table = pd.crosstab(
        df.loc[df['month'] == month, 'city'],
        df.loc[df['month'] == month, 'any_precipitation']
    )
    # Calculate the number of true values (days with precipitation) for each city

    days_with_precipitation = contingency_table[True]

    # Calculate the total number of days for each city
    total_counts = contingency_table.sum(axis=1)

    # Hypothesis test
    ztat, p_value = proportions_ztest(
        count=days_with_precipitation,
        nobs=total_counts,
        alternative='two-sided'
        )
    
    if p_value < significance_level:
        significance_difference[month - 1] = 1
    
    print(f"Month {month}:")
    print(f" z-statistic: {ztat:.2f}")
    print(f" p-value: {p_value:.3f}")
    print("-" * 20)
```
